# Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(s: str) -> int:
    
    vowels = "aeiou"
    return sum(1 for char in s.lower() if char in vowels)


def test_count_vowels():
    assert count_vowels("hello") == 2      # e, o
    assert count_vowels("world") == 1      # o
    assert count_vowels("Python") == 1     # o
    assert count_vowels("AEIOU") == 5      # все гласные
    assert count_vowels("bcdfg") == 0      # нет гласных
    assert count_vowels("") == 0           # пустая строка
    assert count_vowels("Communication") == 6  # o,u,i,a,i,o

    print("Все тесты прошли успешно!")


test_count_vowels()
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def has_unique_chars(s: str) -> bool:
    
    return len(set(s)) == len(s)

def test_has_unique_chars():
    assert has_unique_chars("abcde") is True          # все уникальные
    assert has_unique_chars("hello") is False          # 'l' повторяется
    assert has_unique_chars("") is True                # пустая строка — уникальна
    assert has_unique_chars("a") is True               # одна буква
    assert has_unique_chars("Aa") is True              # разные символы (регистр важен)
    assert has_unique_chars("1123") is False           # '1' повторяется
    assert has_unique_chars("abcdefg!") is True        # с символами тоже работает

    print("Все тесты прошли успешно!")

test_has_unique_chars()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_bits(n: int) -> int:

    if n < 0:
        raise ValueError("Число должно быть положительным.")
    return bin(n).count("1")

def test_count_bits():
    assert count_bits(0) == 0              # 0 -> 0b0
    assert count_bits(1) == 1              # 1 -> 0b1
    assert count_bits(5) == 2              # 5 -> 0b101
    assert count_bits(7) == 3              # 7 -> 0b111
    assert count_bits(10) == 2             # 10 -> 0b1010
    assert count_bits(255) == 8            # 255 -> 0b11111111
    assert count_bits(1024) == 1           # 1024 -> 0b10000000000

    print("Все тесты прошли успешно!")


test_count_bits()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def multiplicative_persistence(n: int) -> int:

    if n < 0:
        raise ValueError("Число должно быть положительным.")
    
    steps = 0
    while n >= 10: 
        product = 1
        for digit in str(n):
            product *= int(digit)
        n = product
        steps += 1
    return steps

def test_multiplicative_persistence():
    assert multiplicative_persistence(39) == 3
    assert multiplicative_persistence(4) == 0
    assert multiplicative_persistence(999) == 4
    assert multiplicative_persistence(25) == 2    # 2*5=10 -> 1*0=0
    assert multiplicative_persistence(77) == 4    # 7*7=49 -> 4*9=36 -> 3*6=18 -> 1*8=8
    assert multiplicative_persistence(10) == 1    # 1*0=0
    assert multiplicative_persistence(0) == 0     # уже однозначное число

    print("Все тесты прошли успешно!")


test_multiplicative_persistence()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
   
    if len(pred) != len(true):
        raise ValueError("Векторы должны быть одинаковой длины.")
    if not pred:  # если векторы пустые
        raise ValueError("Векторы не должны быть пустыми.")
    
    squared_errors = [(p - t) ** 2 for p, t in zip(pred, true)]
    return sum(squared_errors) / len(pred)

def test_mse():
    assert mse([1, 2, 3], [1, 2, 3]) == 0.0                   # совпадают
    assert mse([1, 2, 3], [2, 3, 4]) == 1.0                   # (1+1+1)/3
    assert mse([10, 20, 30], [10, 10, 10]) == ((0**2 + 10**2 + 20**2)/3)
    assert round(mse([0, 0, 0], [1, 1, 1]), 2) == 1.0         # (1+1+1)/3
    assert mse([-1, -2, -3], [-1, -2, -3]) == 0.0             # отрицательные значения

    try:
        mse([1, 2], [1])  
    except ValueError:
        pass
    else:
        assert False, "Ожидалось исключение при разных длинах векторов"

    try:
        mse([], [])
    except ValueError:
        pass
    else:
        assert False, "Ожидалось исключение при пустых векторах"

    print("Все тесты прошли успешно!")



test_mse()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors(n: int) -> str:
  
    if n <= 0:
        raise ValueError("Число должно быть положительным.")

    factors = {}
    divisor = 2

    while n > 1:
        count = 0
        while n % divisor == 0:
            n //= divisor
            count += 1
        if count > 0:
            factors[divisor] = count
        divisor += 1

    # Формируем строку вывода
    result = ""
    for prime in sorted(factors.keys()):
        power = factors[prime]
        if power == 1:
            result += f"({prime})"
        else:
            result += f"({prime}**{power})"

    return result

def test_prime_factors():
    assert prime_factors(86240) == "(2**5)(5)(7**2)(11)"
    assert prime_factors(7775460) == "(2**2)(3**3)(5)(7)(11**2)(17)"
    assert prime_factors(17) == "(17)"
    assert prime_factors(360) == "(2**3)(3**2)(5)"
    assert prime_factors(1) == ""  # у 1 нет простых множителей

    try:
        prime_factors(0)
    except ValueError:
        pass
    else:
        assert False, "Ожидалось исключение при n <= 0"

    print("Все тесты прошли успешно!")



test_prime_factors()
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    if number <= 0:
        return "It is impossible"

    total = 0
    k = 0
    while total < number:
        k += 1
        total += k ** 2

    if total == number:
        return k
    else:
        return "It is impossible"


def test_pyramid():
    assert pyramid(14) == 3         # 1^2 + 2^2 + 3^2 = 14
    assert pyramid(5) == 2          # 1^2 + 2^2 = 5
    assert pyramid(1) == 1          # 1^2 = 1
    assert pyramid(30) == 4         # 1^2+2^2+3^2+4^2=30
    assert pyramid(55) == 5         # 1+4+9+16+25=55
    assert pyramid(0) == "It is impossible"
    assert pyramid(-10) == "It is impossible"

    print("Все тесты прошли успешно!")

test_pyramid()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(number: int) -> bool:
 
    if number <= 0:
        raise ValueError("Число должно быть положительным.")

    digits = [int(d) for d in str(number)]
    n = len(digits)

    if n == 1:
        return True  

    if n % 2 == 0:
        mid_left = n//2 - 1
        mid_right = n//2
        left_sum = sum(digits[:mid_left])
        right_sum = sum(digits[mid_right+1:])
    else:
        mid = n//2
        left_sum = sum(digits[:mid])
        right_sum = sum(digits[mid+1:])

    return left_sum == right_sum

def test_is_balanced():
    assert is_balanced(123803) == True    # 12|38|03 -> 1+2=0+3? нужно проверить
    assert is_balanced(123803) == True    # 123803: digits=[1,2,3,8,0,3], even, middle 3,8, left_sum=1+2=3, right_sum=0+3=3 -> True
    assert is_balanced(12345) == False    # 1+2 != 4+5
    assert is_balanced(12321) == True     # 1+2=2+1
    assert is_balanced(7) == True         # однозначное число
    assert is_balanced(1423) == False
    assert is_balanced(123321) == True    # 12|33|21 -> 1+2=2+1=3? left_sum=1+2=3, right_sum=2+1=3 -> True

    try:
        is_balanced(0)
    except ValueError:
        pass
    else:
        assert False, "Ожидалось исключение при number <= 0"

    print("Все тесты прошли успешно!")

test_is_balanced()
```
